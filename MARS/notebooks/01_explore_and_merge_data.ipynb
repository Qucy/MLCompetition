```python
import os
import pandas as pd


training_data_path = '../智能制造训练集/train/inputs/'
training_label_file = '../智能制造训练集/train/outputs/主蒸汽流量.csv'
testing_data_path = '../智能制造赛道测试集/test/'
```

## 1 - loading and merge training data and labels

```python
%%time

# init dataframe
x_train = pd.DataFrame()
y_train = pd.DataFrame()
x_test = pd.DataFrame()

# looping training input files and merge into one big file
training_files = os.listdir(training_data_path)

for file_name in training_files:
    # skip training file
    if file_name == 'training.csv':
        continue
    # concat file path and file name into a relative path
    file_path = os.path.join(training_data_path, file_name)
    # loading csv file
    ts_feature = pd.read_csv(file_path).drop_duplicates()
    # log
    print(f"Loaded file {file_path} and file shape is {ts_feature.shape}")
    # concat
    if len(x_train) == 0:
        x_train = ts_feature
    else:
        x_train = pd.merge(x_train, ts_feature, on=['时间'], how='left')
```

```python
# check missing data point
full_date_range = pd.date_range(start='2021-12-20 00:00:00', end='2021-12-22 23:59:59', freq='S')

full_date_range = [str(t) for t in full_date_range]

training_data_missing_time_points = set(full_date_range).difference(set(x_train['时间'].unique()))

len(training_data_missing_time_points)
```

## 4 - Merge and save testing data

```python
# looping training input files and merge into one big file
training_files = os.listdir(testing_data_path)

for file_name in training_files:
    # skip testing file
    if file_name == 'testing.csv':
        continue
    # concat file path and file name into a relative path
    file_path = os.path.join(testing_data_path, file_name)
    # loading csv file
    ts_feature = pd.read_csv(file_path).drop_duplicates()
    # log
    print(f"Loaded file {file_path} and file shape is {ts_feature.shape}")
    # concat
    if len(x_test) == 0:
        x_test = ts_feature
    else:
        x_test = pd.merge(x_test, ts_feature, on=['时间'], how='left')
```
